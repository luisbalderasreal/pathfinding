```python
import networkx as nx
import numpy as np
from numpy import linalg as ln
from scipy.spatial import Delaunay
from random import randint

points = np.empty([0,2])
while len(points)<100:
    x = randint(0,99)
    y = randint(0,99)
    if [x,y] not in points:
        points=np.append(points, [[x, y]],axis=0)
tri=Delaunay(points)
(points,tri.simplices)
```

```python
import matplotlib.pyplot as plt
plt.triplot(points[:,0], points[:,1], tri.simplices)
plt.plot(points[:,0], points[:,1], 'o')
plt.show()
```

```python
G=nx.Graph()
for path in tri.simplices:
    G.add_edge(path[0],path[1],weight=ln.norm(points[path[0]]-points[path[1]]))
    G.add_edge(path[1],path[2],weight=ln.norm(points[path[1]]-points[path[2]]))
    G.add_edge(path[2],path[0],weight=ln.norm(points[path[2]]-points[path[0]]))
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G,pos)
nx.draw_networkx_edges(G, pos,G.edges())
nx.draw_networkx_labels(G, pos)
edge_labels = nx.get_edge_attributes(G, "weight")
nx.draw_networkx_edge_labels(G, pos, edge_labels)
```
