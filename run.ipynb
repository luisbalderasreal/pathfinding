```python
from random import randint

size = 16
obs_count = 150

obstacle_count = 0
obstacles = []
while obstacle_count < obs_count:
    obstacle = (randint(0,size-1),randint(0,size-1))
    if obstacle not in obstacles and obstacle != (0,0) and obstacle != (size-1,size-1):
        obstacles.append(obstacle)
        obstacle_count += 1
```

```python
B = AStarSolver(Knight(obstacles,16,16))
```

```python
start = time.time()
B.solve((0,0),(15,15))
end = time.time()
print(end-start)
```

```python
A = IDAStarSolver(Knight(obstacles,16,16))
```

```python
start = time.time()
A.solve((0,0),(15,15))
end = time.time()
print(end-start)
```
